# Red neuronal
Sea $x_{1}$ y $x_{2}$ variables  $x_{1},x_{2}\in \mathcal{N}(0,1)$
$y=sign(x_{1}*w_{1}+x_{2}*w_{2}+w_{0})$
donde $w_{1}=-3.84, w_{2}=20.73, w_{0}=10$

```python
w1 = -3.84
w2 = 20.73
w0 = 10

x1 = np.random.randn(1000)
x2 = np.random.randn(1000)
y = np.sign(x1*w1 + x2*w2 + w0)
y = y*0.5+0.5
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter3d(x=x1, y=x2, z=y, mode='markers'))
```

```python
def logsig(z):
  return 1./(1+np.exp(-z))

def J(yreal,ypred):
  error = (yreal - ypred)**2
  meanError = np.mean(error)
  return error, meanError

def dJ_dw1(yreal,ypred,x1):
  grad = (yreal - ypred)*(-1)*(ypred*(1-ypred))*x1
  meanGrad = np.mean(grad)
  return grad,meanGrad

def dJ_dw2(yreal,ypred,x2):
  grad = (yreal - ypred)*(-1)*(ypred*(1-ypred))*x2
  meanGrad = np.mean(grad)
  return grad,meanGrad

def dJ_dw0(yreal,ypred):
  grad = (yreal - ypred)*(-1)*(ypred*(1-ypred))*1
  meanGrad = np.mean(grad)
  return grad,meanGrad
```

```python
mu = 2.5
we1 = 2.3
we2 = 2.3
we0 = 1.0
niter = 10000

ERROR=[]
MTX_we1=[]
MTX_we2=[]
MTX_we0=[]
for i in range(niter):
  ypred = logsig(we1*x1 + we2*x2 + we0)
  error, meanError  = J(y, ypred)
  ERROR.append(meanError)
  _, dw1 = dJ_dw1(y, ypred, x1)
  we1 = we1 - mu*dw1
  _, dw2 = dJ_dw2(y, ypred, x2)
  we2 = we2 - mu*dw2
  _, dw0 = dJ_dw0(y, ypred)
  we0 = we0 - mu*dw0
  MTX_we1.append(we1)
  MTX_we2.append(we2)
  MTX_we0.append(we0)
  # print(i)
```

```python
fig=go.Figure()
fig.add_trace(go.Scatter(y=np.array(MTX_we1),name='w1'))
fig.add_trace(go.Scatter(y=np.array(MTX_we2),name='w2'))
fig.add_trace(go.Scatter(y=np.array(MTX_we0),name='w0'))
fig.add_trace(go.Scatter(y=np.array(ERROR),name='ERROR'))
```
