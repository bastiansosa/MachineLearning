```python
X = np.array([[20,10],[17,22]])
Y = np.array([[350],[500]])
```

```python
theta =np.linalg.inv(X.T @ X) @ (X.T @ Y)
theta
```

```python
np.array([[3,8],[4,30]]) @ theta
```

```python
TT = np.array([[-9],[1]])
mu = 0.001
for i in range(1000):
  TT = TT + mu*((X.T @ Y) - (X.T@X)@TT)
  print('iter:%2d - x:%2.4f, y:%2.4f'%(i,TT[0],TT[1]))
```
